### Q4: Write a function to get the scorecard of each match. This function would take the innings value as argument and return two scorecard dataframes each for one team as shown below. So the first dataframe would contain the top 4 scorers of the team who batted first and the top 4 bowlers of the opponent team. And the second dataframe would contain the top 4 scorers of the team who batted second and the top 4 bowlers of the opponent team.

```python
def over(a):
    return int(a/6)+(a%6)/10
```

```python
def scorecard_per_innings(innings):
    
    batsmen = {x[list(x.keys())[0]]['batsman'] for i, x in enumerate(innings)}
    bowler = {x[list(x.keys())[0]]['bowler'] for i, x in enumerate(innings)}

    dfs1=pd.DataFrame(batsmen)
    dfs1['runs_taken']=0
    dfs1['balls']=0
    dfs1.rename(columns={0:'Batsmen'}, inplace=True)

    dfs2 = pd.DataFrame(bowler)
    dfs2['runs_given']=0
    dfs2['overs']=0
    dfs2['wickets']=0
    dfs2.rename(columns={0:'Bowler'}, inplace=True)

    total_runs = 0

    for i, x in enumerate(innings):
        run = x[list(x.keys())[0]]['runs']['batsman']
        extra = x[list(x.keys())[0]]['runs']['extras']
        total_runs += x[list(x.keys())[0]]['runs']['total']
        if extra!=0:
            extratype = list(x[list(x.keys())[0]]['extras'].keys())[0]
        dfs1.loc[dfs1['Batsmen']==x[list(x.keys())[0]]['batsman'], 'runs_taken']+=run
        if extra==0 or (extra!=0 and (extratype in ['legbyes', 'byes', 'noballs'])):
            dfs1.loc[dfs1['Batsmen']==x[list(x.keys())[0]]['batsman'], 'balls']+=1

    for i, x in enumerate(innings):
        extra = x[list(x.keys())[0]]['runs']['extras']
        runs = x[list(x.keys())[0]]['runs']['total']
        if 'wicket' in list(x[list(x.keys())[0]].keys()):
            dfs2.loc[dfs2['Bowler']==x[list(x.keys())[0]]['bowler'], 'wickets']+=1
        extratype=0
        if extra!=0:
            extratype = list(x[list(x.keys())[0]]['extras'].keys())[0]
        if extra==0 or (extra!=0 and (extratype in ['legbyes', 'byes'])):
            dfs2.loc[dfs2['Bowler']==x[list(x.keys())[0]]['bowler'], 'overs']+=1
        if extratype!='legbyes' and extratype!='byes':
            dfs2.loc[dfs2['Bowler']==x[list(x.keys())[0]]['bowler'], 'runs_given']+=runs

    dfs2['overs']=dfs2['overs'].apply(lambda x: over(x))
    
    dfs1.sort_values(['runs_taken', 'balls'], ascending = [False, True], inplace=True)
    dfs1.reset_index(drop=True, inplace=True)
    dfs2.sort_values(['wickets', 'runs_given', 'overs'], ascending=[False, True, True], inplace=True)
    dfs2.reset_index(drop=True, inplace=True)
    dfs2 = dfs2[['Bowler', 'overs', 'runs_given', 'wickets']]

    total_overs = dfs2['overs'].sum()
    total_wickets = dfs2['wickets'].sum()
    
    return [dfs1, dfs2, total_runs, total_wickets, total_overs]
```

```python
def display_board(sc1, sc2, team):
    print(f"First innings:\n\t{team[0]}: Total score = {sc1[2]}-{sc1[3]} in {sc1[4]} overs")
    print(f"{sc1[0].to_string(index=False)}\n\n\t{team[1]}\n{sc1[1].to_string(index=False)}")
    print(f"\n\nSecond innings:\n\t{team[1]}: Total score = {sc2[2]}-{sc2[3]} in {sc2[4]} overs")
    print(f"{sc2[0].to_string(index=False)}\n\n\t{team[0]}\n{sc2[1].to_string(index=False)}")
```

```python
def scorecard(match):
    innings_1 = eval(match)[0]['1st innings']
    innings_2 = eval(match)[1]['2nd innings']
    a_scorecard = scorecard_per_innings(innings_1['deliveries'])
    b_scorecard = scorecard_per_innings(innings_2['deliveries'])
    teams = [innings_1['team'], innings_2['team']]
    display_board(a_scorecard, b_scorecard, teams)
```

```python
import time
a = time.time()
match_row = 12
scorecard(df.loc[match_row, 'innings'])
try:
    print(f"\n\n{df.loc[match_row, 'info.outcome.winner']} won by {int(df.loc[match_row, 'info.outcome.by.wickets'])} wickets")
except:
    print(f"\n\n{df.loc[match_row, 'info.outcome.winner']} won by {int(df.loc[match_row, 'info.outcome.by.runs'])} runs")
b = time.time()
b-a
```
